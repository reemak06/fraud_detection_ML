```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pathlib import Path
from collections import OrderedDict

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score,
    roc_curve, precision_recall_fscore_support
)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Optional: VIF for multicollinearity (continuous features only)
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Optional: SMOTE if available
try:
    from imblearn.over_sampling import SMOTE
    HAS_SMOTE = True
except Exception:
    HAS_SMOTE = False

# -------------------------------
# Paths
# -------------------------------
DATA_PATH = "/Users/reemakdawe/accredian_fraud_project/data/Fraud.csv"
OUT_DIR = Path("/Users/reemakdawe/accredian_fraud_project/outputs")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# -------------------------------
# 1) LOAD
# -------------------------------
df = pd.read_csv(DATA_PATH)
print("Loaded shape:", df.shape)
print(df.head(3))

# -------------------------------
# Quick sanity for target
# -------------------------------
TARGET = "isFraud"
if TARGET not in df.columns:
    raise ValueError(f"Target column '{TARGET}' not found. Columns: {df.columns.tolist()}")

print("\nTarget distribution:")
print(df[TARGET].value_counts(dropna=False))

if df[TARGET].nunique() < 2:
    # Cannot train with only one class
    msg = (
        f"\n Your dataset contains only one class in '{TARGET}'.\n"
        f"value_counts:\n{df[TARGET].value_counts(dropna=False)}\n\n"
        " Fix: Load the full dataset or ensure fraud rows (isFraud=1) are present.\n"
        "Once both classes exist, re-run this script."
    )
    print(msg)
    # Save a note and stop
    (OUT_DIR / "STOP_ONE_CLASS_ONLY.txt").write_text(msg)
    raise SystemExit

# -------------------------------
# 1) DATA CLEANING
#    - Missing values
#    - Outliers (winsorize 1%/99%)
#    - Multicollinearity (corr + VIF)
# -------------------------------

# Drop obvious ID-like columns if present
id_like = ["nameOrig", "nameDest", "step_id", "transaction_id"]
for c in id_like:
    if c in df.columns:
        df.drop(columns=c, inplace=True)

# Basic report of missingness
missing_report = df.isnull().sum().sort_values(ascending=False)
missing_report.to_csv(OUT_DIR / "missing_report.csv")
print("\nMissing report saved to outputs/missing_report.csv")

# Simple imputation strategy:
# - Numeric: median
# - Categorical: most frequent
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()
if TARGET in num_cols:
    num_cols.remove(TARGET)
if TARGET in cat_cols:
    cat_cols.remove(TARGET)

# Winsorize (cap) outliers on numeric columns (except target)
def winsorize(df_, cols, lower_q=0.01, upper_q=0.99):
    df_w = df_.copy()
    for c in cols:
        ql, qu = df_w[c].quantile([lower_q, upper_q])
        df_w[c] = df_w[c].clip(lower=ql, upper=qu)
    return df_w

df[num_cols] = winsorize(df, num_cols)[num_cols]

# -------------------------------
# Multicollinearity (Pearson corr on numeric)
# -------------------------------
corr = df[num_cols + [TARGET]].corr(numeric_only=True)
corr.to_csv(OUT_DIR / "correlation_matrix.csv")

# Drop highly correlated predictors (> 0.95) pairwise
high_corr_thresh = 0.95
to_drop_corr = set()
corr_mat = corr.copy()
for i, c1 in enumerate(num_cols):
    for j, c2 in enumerate(num_cols):
        if j <= i:
            continue
        val = abs(corr_mat.loc[c1, c2])
        if val > high_corr_thresh:
            # drop the one that is more correlated with others
            to_drop_corr.add(c2)

kept_num_cols = [c for c in num_cols if c not in to_drop_corr]
print("\nDropped (high corr) numeric columns:", sorted(list(to_drop_corr)))

# -------------------------------
# VIF on continuous features (optional, robust check)
# Only run VIF on kept_num_cols to avoid singularities
# -------------------------------
def compute_vif(df_, cols):
    if len(cols) < 2:
        return pd.DataFrame({"feature": cols, "VIF": [np.nan]*len(cols)})
    X = df_[cols].astype(float).values
    vif_data = []
    for i, c in enumerate(cols):
        try:
            vif_val = variance_inflation_factor(X, i)
        except Exception:
            vif_val = np.nan
        vif_data.append((c, vif_val))
    return pd.DataFrame(vif_data, columns=["feature", "VIF"]).sort_values("VIF", ascending=False)

vif_df = compute_vif(df, kept_num_cols)
vif_df.to_csv(OUT_DIR / "vif_report.csv", index=False)
print("VIF report saved to outputs/vif_report.csv")

# Optionally drop VIF>20 columns (if any)
vif_drop = vif_df.loc[vif_df["VIF"] > 20, "feature"].tolist()
print("VIF>20 columns to drop:", vif_drop)
kept_num_cols = [c for c in kept_num_cols if c not in vif_drop]

# -------------------------------
# 2) & 3) MODEL + VARIABLE SELECTION (Pipeline)
# - OneHot for categoricals
# - Median/Most-freq imputation
# - RandomForest with class weights
# -------------------------------

# Rebuild lists after drops
predictors_num = kept_num_cols
predictors_cat = [c for c in cat_cols if c in df.columns]

X = df[predictors_num + predictors_cat]
y = df[TARGET]

# Preprocessor
numeric_tf = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

cat_tf = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore", drop="first"))
])

preproc = ColumnTransformer(
    transformers=[
        ("num", numeric_tf, predictors_num),
        ("cat", cat_tf, predictors_cat)
    ],
    remainder="drop"
)

# Base classifier
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=None,
    n_jobs=-1,
    random_state=42,
    class_weight="balanced_subsample"
)

pipeline = Pipeline(steps=[
    ("preproc", preproc),
    ("clf", rf)
])

# -------------------------------
# 4) SPLIT + (Optional) SMOTE only on train
# -------------------------------
X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Fit preprocessor only to get encoded numeric matrix for SMOTE (if we use it)
preproc_only = preproc.fit(X_train)
X_train_enc = preproc_only.transform(X_train)

# Use SMOTE only if:
# - library available
# - both classes exist in y_train
use_smote = False
if HAS_SMOTE and y_train.nunique() == 2:
    # Check minority count
    counts = y_train.value_counts()
    if counts.min() >= 10:  # at least some positives to synthesize
        use_smote = True

if use_smote:
    sm = SMOTE(sampling_strategy="minority", random_state=42)
    Xs, ys = sm.fit_resample(X_train_enc, y_train)
    print("\nSMOTE applied. New train distribution:\n", ys.value_counts())
    # Rebuild a pipeline that skips preproc during .fit (we already transformed)
    # But for simplicity, well keep the same pipeline and just fit classifier manually:
    clf = rf
    clf.fit(Xs, ys)
    # For validation, we must transform via preproc_only
    X_valid_enc = preproc_only.transform(X_valid)
    y_pred = clf.predict(X_valid_enc)
    y_scores = clf.predict_proba(X_valid_enc)[:, 1]
else:
    print("\nSMOTE not applied (either not installed or too few minority samples).")
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_valid)
    if hasattr(pipeline.named_steps["clf"], "predict_proba"):
        y_scores = pipeline.predict_proba(X_valid)[:, 1]
    else:
        # Fallback score via decision_function not typical for RF
        y_scores = None

# -------------------------------
# 4) PERFORMANCE REPORTS
# -------------------------------
cm = confusion_matrix(y_valid, y_pred)
report = classification_report(y_valid, y_pred, digits=4)
try:
    roc_auc = roc_auc_score(y_valid, y_scores) if y_scores is not None else np.nan
except Exception:
    roc_auc = np.nan

metrics_txt = OUT_DIR / "metrics.txt"
with metrics_txt.open("w") as f:
    f.write("=== Classification Report ===\n")
    f.write(report + "\n")
    f.write(f"ROC-AUC: {roc_auc:.6f}\n")
print("\nSaved metrics to outputs/metrics.txt")
print(report)
print("ROC-AUC:", roc_auc)

# Confusion matrix plot
plt.figure(figsize=(4,4))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
for (i, j), v in np.ndenumerate(cm):
    plt.text(j, i, str(v), ha="center", va="center")
plt.tight_layout()
plt.savefig(OUT_DIR / "confusion_matrix.png", dpi=150)
plt.close()

# ROC curve
if y_scores is not None and not np.isnan(roc_auc):
    fpr, tpr, thr = roc_curve(y_valid, y_scores)
    plt.figure(figsize=(5,4))
    plt.plot(fpr, tpr, label=f"ROC AUC = {roc_auc:.3f}")
    plt.plot([0,1],[0,1],"--")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC Curve")
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.savefig(OUT_DIR / "roc_curve.png", dpi=150)
    plt.close()

# -------------------------------
# 5) FEATURE IMPORTANCE
# -------------------------------
def rf_feature_importance(fitted_pipeline, preproc_fitted, clf_fitted, X_sample):
    # Get feature names after preprocessing
    num_names = preproc_fitted.named_transformers_["num"].get_feature_names_out(predictors_num)
    cat_names = preproc_fitted.named_transformers_["cat"].named_steps["ohe"].get_feature_names_out(predictors_cat)
    feat_names = np.concatenate([num_names, cat_names])
    importances = clf_fitted.feature_importances_
    return pd.DataFrame({"feature": feat_names, "importance": importances}).sort_values(
        "importance", ascending=False
    )

if use_smote:
    feat_imp_df = rf_feature_importance(None, preproc_only, clf, X_train_enc)
else:
    preproc_fitted = pipeline.named_steps["preproc"]
    clf_fitted = pipeline.named_steps["clf"]
    feat_imp_df = rf_feature_importance(pipeline, preproc_fitted, clf_fitted, X_train)

feat_imp_df.to_csv(OUT_DIR / "feature_importance.csv", index=False)
print("Saved feature importance to outputs/feature_importance.csv")

# -------------------------------
# 6) INTERPRETABILITY HINTS (Top Features)
# -------------------------------
top_factors = feat_imp_df.head(10)
print("\nTop predictive factors:")
print(top_factors)

# -------------------------------
# 7) PREVENTION RECOMMENDATIONS (saved text)
# -------------------------------
prevention = """\
Prevention Actions:
1) Real-time rules for TRANSFER/CASH_OUT with unusually high amounts or zero oldbalanceOrg.
2) Velocity checks: multiple transfers within short intervals  flag.
3) Strong customer auth (OTP/2FA), especially for new payees or large-value transfers.
4) Behavioral profiling: compare to customers historical ranges (amount, time-of-day, recipient novelty).
5) Destination account risk scoring (age of account, prior suspicious inflows).
6) Step-up verification on high-risk signals (device change, IP change, first-time beneficiary).
"""
(OUT_DIR / "prevention_recommendations.txt").write_text(prevention)

# -------------------------------
# 8) HOW TO MEASURE IMPACT (saved text)
# -------------------------------
measure = """\
Effectiveness Measurement (post-implementation):
- Before/After fraud loss (total $ and rate).
- Recall@K of alerts (how many frauds caught among top K alerts).
- False positive rate (to protect customer experience).
- Average time-to-detection; % stopped pre-settlement.
- A/B test (pilot group vs control): compare fraud rate, customer friction.
"""
(OUT_DIR / "effectiveness_measurement.txt").write_text(measure)

# -------------------------------
# Save validation predictions
# -------------------------------
if use_smote:
    Xv_enc = preproc_only.transform(X_valid)
    y_prob = clf.predict_proba(Xv_enc)[:, 1]
else:
    y_prob = pipeline.predict_proba(X_valid)[:, 1]

scored = X_valid.copy()
scored[TARGET] = y_valid.values
scored["fraud_probability"] = y_prob
scored.to_csv(OUT_DIR / "validation_scored.csv", index=False)
print("Saved scored validation to outputs/validation_scored.csv")

print("\n Done. Open the 'outputs' folder for:")
print("- metrics.txt, confusion_matrix.png, roc_curve.png")
print("- feature_importance.csv (Q5 & Q6)")
print("- prevention_recommendations.txt (Q7)")
print("- effectiveness_measurement.txt (Q8)")
print("- correlation_matrix.csv, vif_report.csv (Q1)")
print("- validation_scored.csv (Q4)")
```
